# PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
pca.fit(ds)
```

```python
pca.n_components_
```

```python
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```

```python
plt.figure(figsize=(100, 100))
plt.xticks(np.arange(1, 600, 1))
plt.grid(b=3)
plt.plot(pca.explained_variance_ratio_)
```

```python
pca2 = PCA(n_components=2)
```

```python
for col in ds.columns[:2]:
    sns.distplot(ds[col])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score,confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.feature_selection import RFECV
```

```python
labels
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(ds, labels, test_size=0.3)
```

```python
transTrain = pca2.fit_transform(xTrain)
```

```python
transTest = pca2.transform(xTest)
```

```python
yTrain
```

```python
clasT = RandomForestClassifier()
clasT = clasT.fit(transTrain, yTrain)
```

```python
ac = accuracy_score(yTest, clasT.predict(transTest))
```

```python
ac
```

```python
cm = confusion_matrix(yTest, clasT.predict(transTest))
sns.heatmap(cm, annot = True)
```
