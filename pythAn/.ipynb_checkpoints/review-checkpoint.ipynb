# Machine Learning

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score,confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.feature_selection import RFECV
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
clas1 = RandomForestClassifier()

rfecv = RFECV(clas1, cv=5, scoring='accuracy')
rfecv.fit(xTrain, yTrain)
```

```python
rfecv.n_features_
```

```python
xTest2 = rfecv.transform(xTest)
xTrain2 = rfecv.transform(xTrain)
```

probujemy uczyc na tym co dostalismy

```python
clasT = RandomForestClassifier()

clasT = clasT.fit(xTrain2, yTrain)

ac = accuracy_score(yTest, clasT.predict(xTest2))

print(ac)

cm2 = confusion_matrix(yTest, clasT.predict(xTest2))

sns.heatmap(cm2, annot=True)
                       
    
```

# PCA

```python
from sklearn.decomposition import PCA

xTrainNorm = (xTrain - xTrain.mean() ) / xTrain.std()

pca = PCA()
pca.fit(x)
```

```python
pca.explained_variance_
pca.n_components_
```

inne podejscie do wyboru kilku

```python
from sklearn.feature_selection import RFE
```

```python
clas = RandomForestClassifier()
rfe = RFE(clas, n_features_to_select=4, step=1)
rfe = rfe.fit(xTrain, yTrain)
```

```python
xTrain.columns[rfe.support_]
```

```python
xTrainCorr = rfe.transform(xTrain)
xTestCorr = rfe.transform(xTest)
```

```python
clasN = RandomForestClassifier()
clasN = clasN.fit(xTrainCorr, yTrain)

ac = accuracy_score(yTest, clasN.predict(xTestCorr))
ac
```

```python
cm = confusion_matrix(yTest, clasN.predict(xTestCorr))
sns.heatmap(cm, annot=True)
```
